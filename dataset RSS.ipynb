```python
import librosa

# Definir una función para extraer características (MFCCs)
def extraer_caracteristicas(audio_path):
    audio, sr = librosa.load(audio_path, sr=None)  # Cargar el audio
    mfccs = librosa.feature.mfcc(audio, sr=sr)    # Extraer MFCCs
    return mfccs
```
